# Seaborn

## Distribution plots
- distplot
- joinplot
- pairplot

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind="hex")
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind="reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue="smoker")
```

```python
sns.histplot(df['tip'])
```
