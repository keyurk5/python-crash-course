```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.xlabel("X Axis Title")
plt.ylabel("Y Axis Title")
plt.title("Plot title");
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(y,x,'g')

plt.subplot(1,3,3)
plt.plot(x,y*x,'b')
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.scatter(df['Sepal.Length'], df['Petal.Length'])

plt.subplot(1,3,2)
plt.scatter(df['Petal.Width'], df['Petal.Length'])

plt.subplot(1,3,3)
plt.scatter(df['Sepal.Length'], df['Sepal.Width'])
```

```python
plt.scatter(df['Sepal.Length'], df['Sepal.Width'],alpha=0.5)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.bar(df['Species'] , df['Sepal.Length'])

plt.subplot(1,2,2)
plt.barh(df['Species'] , df['Sepal.Length'], color="red")
```

```python
plt.bar(df['Species'] , df['Sepal.Length'])
plt.xticks(rotation=30, horizontalalignment="center")
```

```python
colours = {"setosa": "#273c75", "versicolor": "#44bd32", "virginica": "#ffbd32"}
plt.bar(df['Species'] , df['Sepal.Length'], color=df['Species'].replace(colours))
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels = mylabels);
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,0.1]

plt.pie(y,labels = mylabels, explode=myexplode);
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,0.1]

plt.pie(y,labels = mylabels, explode=myexplode);
plt.legend()
```
