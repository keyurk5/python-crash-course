# Seaborn

## Distribution plots
- distplot
- joinplot
- pairplot

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind="hex")
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind="reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue="smoker")
```

```python
sns.histplot(df['tip'])
```

## Categorical Plots
- Box plots
- violin plots
- countplot
- bar plot

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot('smoker', data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
sns.barplot(x="total_bill",y="sex",data=df)
```

```python
sns.barplot(y="total_bill",x="smoker",data=df)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette="rainbow")
```

```python
sns.boxplot(x='day',y='total_bill',data=df,hue="sex",palette="rainbow")
```

### Violinplot - combination of boxplot and distribution

```python
sns.violinplot(x='day',y='total_bill',data=df,hue="sex",palette="rainbow")
```
